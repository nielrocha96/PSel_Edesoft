```python
from tkinter import *; from tkinter import ttk; import sqlite3
from tkinter import tix
root=tix.Tk()

class Funcs():
    def variaveis(self):
        self.ID = self.ID_entry.get()
        self.Originador = self.Originador_entry.get()
        self.DocOriginador = self.DocOriginador_entry.get()
        self.Cedente = self.Cedente_entry.get()
        self.DocCedente = self.DocCedente_entry.get()
        self.CCB = self.CCB_entry.get()
        self.Cliente = self.Cliente_entry.get()
        self.CPFCNPJ = self.CPFCNPJ_entry.get()
        self.Endereco = self.Endereco_entry.get()
        self.CEP = self.CEP_entry.get()
        self.Cidade = self.Cidade_entry.get()
        self.UF = self.UF_entry.get()
        self.ValorEmprestimo = self.ValorEmprestimo_entry.get()
        self.ParcelaRS_entry = self.ParcelaRS_entry.get()
        self.TotalParcelas = self.TotalParcelas_entry.get()
        self.Parcela = self.Parcela_entry.get()
        self.DataEmissao = self.DataEmissao_entry.get()
        self.DataVencimento = self.DataVencimento_entry.get()
        self.PrecoAquisicao = self.PrecoAquisicao_entry.get()
        
    def limpa_tela(self):
        self.ID_entry.delete(0, END)
        self.Originador_entry.delete(0, END)
        self.DocOriginador_entry.delete(0, END)
        self.Cedente_entry.delete(0, END)
        self.DocCedente_entry.delete(0, END)
        self.CCB_entry.delete(0, END)
        self.Cliente_entry.delete(0, END)
        self.CPFCNPJ_entry.delete(0, END)
        self.Endereco_entry.delete(0, END)
        self.CEP_entry.delete(0, END)
        Cidade_entry.delete(0, END)
        self.UF_entry.delete(0, END)
        self.ValorEmprestimo_entry.delete(0, END)
        self.ParcelaRS_entry.delete(0, END)
        self.TotalParcelas_entry.delete(0, END)
        self.Parcela_entry.delete(0, END)
        self.DataEmissao_entry.delete(0, END)
        self.DataVencimento_entry.delete(0, END)
        self.PrecoAquisicao_entry.delete(0, END)
        
    def conecta_bd(self):
        self.conn = sqlite3.connect('CESSAO_FUNDO.db')
        self.cursor = self.conn.cursor(); print('Conectando ao banco de dados')
    
    def desconecta_bd(self): self.conn.close(); print('Desconectando ao banco de dados')
    
    def montaTabelas(self):
        self.conecta_bd()
        # Criar tabela
        self.cursor.execute('''CREATE TABLE IF NOT EXISTS CESSAO_FUNDO (
                              ID_CESSAO INT(22) NOT NULL,
                              ORIGINADOR VARCHAR(250) NOT NULL,
                              DOC_ORIGINADOR INT(50) NOT NULL,
                              CEDENTE VARCHAR(250) NOT NULL,
                              DOC_CEDENTE INT(50) NOT NULL,
                              CCB INT(22) NOT NULL,
                              ID_EXTERNO INT(22) NOT NULL,
                              CLIENTE VARCHAR(250) NOT NULL,
                              CPF_CNPJ VARCHAR(50) NOT NULL,
                              ENDERECO VARCHAR(250) NOT NULL,
                              CEP VARCHAR(50) NOT NULL,
                              CIDADE VARCHAR(250) NOT NULL,
                              UF VARCHAR(50) NOT NULL,
                              VALOR_DO_EMPRESTIMO DECIMAL(10,2) NOT NULL,
                              VALOR_PARCELA DECIMAL(10,2) NOT NULL,
                              TOTAL_PARCELAS INT(22) NOT NULL,
                              PARCELA INT(22) NOT NULL,
                              DATA_DE_EMISSAO DATE NOT NULL,
                              DATA_DE_VENCIMENTO DATE NOT NULL,
                              PRECO_DE_AQUISICAO DECIMAL(10,2) NOT NULL,
                              PRIMARY KEY (ID_CESSAO));''')
        self.conn.commit(); print('Banco de dados criado'); self.desconecta_bd()
        
    def add_cliente(self):
        self.variaveis()
        self.conecta_bd()
        self.cursor.execute(""" INSERT INTO CESSAO_FUNDO (ID_EXTERNO, ORIGINADOR, DOC_ORIGINADOR, CEDENTE,
                                                      DOC_CEDENTE, CCB, CLIENTE,CPF_CNPJ, ENDERECO, CEP,
                                                      CIDADE, UF, VALOR_DO_EMPRESTIMO, VALOR_PARCELA,
                                                      TOTAL_PARCELAS, PARCELA, DATA_DE_EMISSAO,
                                                      DATA_DE_VENCIMENTO,PRECO_DE_AQUISICAO)
                        VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)""",
                            (self.ID, self.Originador, self.DocOriginador, self.Cedente, self.DocCedente, self.CCB,
                             self.Cliente, self.CPFCNPJ, self.Endereco, self.CEP, self.Cidade, self.UF,
                             self.ValorEmprestimo, self.ParcelaRS_entry, self.TotalParcelas, self.Parcela,
                             self.DataEmissao, self.DataVencimento, self.PrecoAquisicao))
        self.conn.commit()
        self.desconecta_bd()
        self.select_lista()
        self.limpa_tela()
        
    def select_lista(self):
        self.listaCli.delete(*self.listaCli.get_children())
        self.conecta_bd()
        lista = self.cursor.execute(""" SELECT ID_CESSAO, ID_EXTERNO, ORIGINADOR, DOC_ORIGINADOR, CEDENTE,
                                                      DOC_CEDENTE, CCB, CLIENTE,CPF_CNPJ, ENDERECO, CEP,
                                                      CIDADE, UF, VALOR_DO_EMPRESTIMO, VALOR_PARCELA,
                                                      TOTAL_PARCELAS, PARCELA, DATA_DE_EMISSAO,
                                                      DATA_DE_VENCIMENTO,PRECO_DE_AQUISICAO
                        FROM CESSAO_FUNDO ORDER BY ID_EXTERNO ASC; """)
        for i in lista: self.listaCli.insert("", END, values=i)
        self.desconecta_bd()
        
    def OnDoubleClick(self, event):
        self.limpa_tela()
        self.listaCli.selection()
        for n in self.listaCli.selection():
            col1, col2, col3, col4, col5, col6, col7, col8, col9, col10, col11, col12, col13, col14, col15, col16, col17, col18, col19 = self.listaCli.item(n, 'values')
        
            self.ID_entry.insert(END, col1)
            self.Originador_entry.insert(END, col2)
            self.DocOriginador_entry.insert(END, col3)
            self.Cedente_entry.insert(END, col4)
            self.DocCedente_entry.insert(END, col5)
            self.CCB_entry.insert(END, col6)
            self.Cliente_entry.insert(END, col7)
            self.CPFCNPJ_entry.insert(END, col8)
            self.Endereco_entry.insert(END, col9)
            self.CEP_entry.insert(END, col10)
            Cidade_entry.insert(END, col11)
            self.UF_entry.insert(END, col12)
            self.ValorEmprestimo_entry.insert(END, col13)
            self.ParcelaRS_entry.insert(END, col14)
            self.TotalParcelas_entry.insert(END, col15)
            self.Parcela_entry.insert(END, col16)
            self.DataEmissao_entry.insert(END, col17)
            self.DataVencimento_entry.insert(END, col18)
            self.PrecoAquisicao_entry.insert(END, col19)
    
    def deleta_cliente(self):
        self.variaveis()
        self.conecta_bd()
        self.cursor.execute("""DELETE FROM CESSAO_FUNDO WHERE cod = ? """, (self.ID))
        self.conn.commit()
        self.desconecta_bd()
        self.limpa_tela()
        self.select_lista()
    
    def altera_cliente(self):
        self.variaveis()
        self.conecta_bd()
        
        self.cursor.execute("""ID_EXTERNO = ?, ORIGINADOR = ?, DOC_ORIGINADOR = ?, CEDENTE = ?, DOC_CEDENTE = ?, CCB = ?,
        CLIENTE,CPF_CNPJ = ?, ENDERECO = ?, CEP = ?, CIDADE = ?, UF = ?, VALOR_DO_EMPRESTIMO = ?,
        VALOR_PARCELA = ?, TOTAL_PARCELAS = ?, PARCELA = ?, DATA_DE_EMISSAO = ?,
        DATA_DE_VENCIMENTO = ?,PRECO_DE_AQUISICAO = ? WHERE ID =?""",
                            (self.ID, self.Originador, self.DocOriginador, self.Cedente, self.DocCedente, self.CCB,
                             self.Cliente, self.CPFCNPJ, self.Endereco, self.CEP, self.Cidade, self.UF,
                             self.ValorEmprestimo, self.ParcelaRS_entry, self.TotalParcelas, self.Parcela,
                             self.DataEmissao, self.DataVencimento, self.PrecoAquisicao))
                             
        self.conn.commit()
        self.desconecta_bd()
        self.select_lista()
        self.limpa_cliente()
        
    def busca_cliente(self):
        self.conecta_bd()
        self.listaCli.delete(*self.listaCli.get_children())
        self.nome_entry.insert(END, '%')
        nome = self.nome_entry.get()
        self.cursor.execute(""" SELECT ID_CESSAO, ID_EXTERNO, ORIGINADOR, DOC_ORIGINADOR, CEDENTE,
                                                      DOC_CEDENTE, CCB, CLIENTE,CPF_CNPJ, ENDERECO, CEP,
                                                      CIDADE, UF, VALOR_DO_EMPRESTIMO, VALOR_PARCELA,
                                                      TOTAL_PARCELAS, PARCELA, DATA_DE_EMISSAO,
                                                      DATA_DE_VENCIMENTO,PRECO_DE_AQUISICAO
                                FROM CESSAO_FUNDO WHERE ID_EXTERNO LIKE '%s' ORDER BY ID_EXTERNO ASC """ % nome)
        buscanomeCli = self.cursor.fetchall()
        for i in buscanomeCli: self.listaCli.insert("",END, values=i)
        self.limpa_tela()
        self.desconecta_bd()

class Application(Funcs):
    def __init__(self):
        self.root = root
        ## self.images_base64()
        self.tela(); self.frames_da_tela();
        self.widgets_frame1(); self.lista_frame2();
        self.montaTabelas(); self.select_lista(); self.Menus()
        root.mainloop()
    
    def tela(self):
        self.root.title('Cadastro de Clientes')                                           #Título da janela
        self.root.configure(background='#1e3743')                                         #Cor de fundo da janela
        self.root.geometry('1800x650+0+0'); self.root.resizable(True,True)                #Dimensões iniciais da janela
        #self.root.maxsize(width=900,height=700); self.root.minsize(width=500,height=400) #Tamanhos máximo e mínimo permitidos
    
    def frames_da_tela(self):
        self.frame1 = Frame(self.root, bd=4, bg='#dfe3ee', highlightbackground='#759fe6', highlightthickness=3)
        self.frame1.place(relx=0.02, rely=0.01, relwidth=0.96,relheight=0.66)
        self.frame2 = Frame(self.root, bd=4, bg='#dfe3ee', highlightbackground='#759fe6', highlightthickness=3)
        self.frame2.place(relx=0.02, rely=0.55, relwidth=0.96,relheight=0.43)
    
    def widgets_frame1(self):
        self.abas = ttk.Notebook(self.frame1)
        self.aba1 = Frame(self.abas);
        self.aba1.configure(background='#def3ee');
        self.abas.add(self.aba1, text='Aba 1');      
        self.abas.place(relx=0,rely=0,relwidth=0.98,relheight=0.98)
        
        # Botão limpar
        self.bt_limpar = Button(self.aba1, text='Limpar',bd=2,bg='#107db2',fg='white', activebackgroun='#108ecb', activeforeground='white', font='verdana 8 bold', command=self.limpa_tela)
        self.bt_limpar.place(relx=0.2, rely=0.1, relwidth=0.1, relheight=0.08)
        # Botão buscar
        self.bt_buscar = Button(self.aba1, text='Buscar',bd=2,bg='#107db2',fg='white',font='verdana 8 bold', command = self.busca_cliente)
        self.bt_buscar.place(relx=0.3, rely=0.1, relwidth=0.1, relheight=0.08)
        
        texto_balao_buscar = 'Digite no campo o \'Nome\' do cliente que deseja pesquisar.'
        self.balao_buscar = tix.Balloon(self.aba1)
        self.balao_buscar.bind_widget(self.bt_buscar, balloonmsg = texto_balao_buscar)
                
        # Botão novo
        self.bt_novo = Button(self.aba1, text='Novo',bd=2,bg='#107db2',fg='white',font='verdana 8 bold', command=self.add_cliente)
        self.bt_novo.place(relx=0.6, rely=0.1, relwidth=0.1, relheight=0.08)
        ##self.bt_novo.config(Image=self.imgNovo) #linha para o caso de estilização de botão
        
        # Botão alterar
        self.bt_alterar = Button(self.aba1, text='Alterar',bd=2,bg='#107db2',fg='white',font='verdana 8 bold', command=self.altera_cliente)
        self.bt_alterar.place(relx=0.7, rely=0.1, relwidth=0.1, relheight=0.08)
        # Botão apagar
        self.bt_apagar = Button(self.aba1, text='Apagar',bd=2,bg='#107db2',fg='white',font='verdana 8 bold',command=self.deleta_cliente)
        self.bt_apagar.place(relx=0.8, rely=0.1, relwidth=0.1, relheight=0.08)

        #  Criação da label e entrada do ID
        self.lb_ID = Label(self.aba1, text='ID',bg='#dfe3ee',fg='#107db2'); self.lb_ID.place(relx=0.05,rely=0.05)
        self.ID_entry = Entry(self.aba1); self.ID_entry.place(relx=0.05,rely=0.1,relwidth=0.08)
        #  Criação da label e entrada do Orginador
        self.lb_Originador = Label(self.aba1, text='Originador',bg='#dfe3ee',fg='#107db2'); self.lb_Originador.place(relx=0.05,rely=0.25)
        self.Originador_entry = Entry(self.aba1); self.Originador_entry.place(relx=0.05,rely=0.3,relwidth=0.2)
        #  Criação da label e entrada do Doc Originador
        self.lb_DocOriginador = Label(self.aba1, text='Doc Originador',bg='#dfe3ee',fg='#107db2'); self.lb_DocOriginador.place(relx=0.3,rely=0.25)
        self.DocOriginador_entry = Entry(self.aba1); self.DocOriginador_entry.place(relx=0.3,rely=0.3,relwidth=0.2)
        #  Criação da label e entrada do Cedente
        self.lb_Cedente = Label(self.aba1, text='Cedente',bg='#dfe3ee',fg='#107db2'); self.lb_Cedente.place(relx=0.55,rely=0.25)
        self.Cedente_entry = Entry(self.aba1); self.Cedente_entry.place(relx=0.55,rely=0.3,relwidth=0.2)
        #  Criação da label e entrada do Doc Cedente
        self.lb_DocCedente = Label(self.aba1, text='Doc Cedente',bg='#dfe3ee',fg='#107db2'); self.lb_DocCedente.place(relx=0.8,rely=0.25)
        self.DocCedente_entry = Entry(self.aba1); self.DocCedente_entry.place(relx=0.8,rely=0.3,relwidth=0.175)
        #  Criação da label e entrada do CCB
        self.lb_CCB = Label(self.aba1, text='CCB',bg='#dfe3ee',fg='#107db2'); self.lb_CCB.place(relx=0.05,rely=0.4)
        self.CCB_entry = Entry(self.aba1); self.CCB_entry.place(relx=0.05,rely=0.45,relwidth=0.1)
        #  Criação da label e entrada do Cliente
        self.lb_Cliente = Label(self.aba1, text='Cliente',bg='#dfe3ee',fg='#107db2'); self.lb_Cliente.place(relx=0.175,rely=0.4)
        self.Cliente_entry = Entry(self.aba1); self.Cliente_entry.place(relx=0.175,rely=0.45,relwidth=0.2)
        #  Criação da label e entrada do CPF/CNPJ
        self.lb_CPFCNPJ = Label(self.aba1, text='CPF/CNPJ',bg='#dfe3ee',fg='#107db2'); self.lb_CPFCNPJ.place(relx=0.4,rely=0.4)
        self.CPFCNPJ_entry = Entry(self.aba1); self.CPFCNPJ_entry.place(relx=0.4,rely=0.45,relwidth=0.2)
        #  Criação da label e entrada do Endereço
        self.lb_Endereco = Label(self.aba1, text='Endereço',bg='#dfe3ee',fg='#107db2'); self.lb_Endereco.place(relx=0.625,rely=0.4)
        self.Endereco_entry = Entry(self.aba1); self.Endereco_entry.place(relx=0.625,rely=0.45,relwidth=0.35)
        #  Criação da label e entrada do CEP
        self.lb_CEP = Label(self.aba1, text='CEP',bg='#dfe3ee',fg='#107db2'); self.lb_CEP.place(relx=0.05,rely=0.55)
        self.CEP_entry = Entry(self.aba1); self.CEP_entry.place(relx=0.05,rely=0.6,relwidth=0.1)
        #  Criação da label e entrada da Cidade
        self.lb_Cidade = Label(self.aba1, text='Cidade',bg='#dfe3ee',fg='#107db2'); self.lb_Cidade.place(relx=0.175,rely=0.55)
        self.Cidade_entry = Entry(self.aba1); self.Cidade_entry.place(relx=0.175,rely=0.6,relwidth=0.15)
        #  Criação da label e entrada da UF
        self.lb_UF = Label(self.aba1, text='UF',bg='#dfe3ee',fg='#107db2'); self.lb_UF.place(relx=0.35,rely=0.55)
        self.UF_entry = Entry(self.aba1); self.UF_entry.place(relx=0.35,rely=0.6,relwidth=0.05)
        #  Criação da label e entrada do Valor do Empréstimo
        self.lb_ValorEmprestimo = Label(self.aba1, text='Valor do Empréstimo',bg='#dfe3ee',fg='#107db2'); self.lb_ValorEmprestimo.place(relx=0.425,rely=0.55)
        self.ValorEmprestimo_entry = Entry(self.aba1); self.ValorEmprestimo_entry.place(relx=0.425,rely=0.6,relwidth=0.15)
        #  Criação da label e entrada da Parcela R$
        self.lb_ParcelaRS = Label(self.aba1, text='Parcela R$',bg='#dfe3ee',fg='#107db2'); self.lb_ParcelaRS.place(relx=0.6,rely=0.55)
        self.ParcelaRS_entry = Entry(self.aba1); self.ParcelaRS_entry.place(relx=0.6,rely=0.6,relwidth=0.15)
        #  Criação da label e entrada do Total Parcelas
        self.lb_TotalParcelas = Label(self.aba1, text='Total Parcelas',bg='#dfe3ee',fg='#107db2'); self.lb_TotalParcelas.place(relx=0.775,rely=0.55)
        self.TotalParcelas_entry = Entry(self.aba1); self.TotalParcelas_entry.place(relx=0.775,rely=0.6,relwidth=0.1)
        #  Criação da label e entrada da Parcela
        self.lb_Parcela = Label(self.aba1, text='Parcela',bg='#dfe3ee',fg='#107db2'); self.lb_Parcela.place(relx=0.9,rely=0.55)
        self.Parcela_entry = Entry(self.aba1); self.Parcela_entry.place(relx=0.9,rely=0.6,relwidth=0.08)
        #  Criação da label e entrada da Data de Emissão
        self.lb_DataEmissao = Label(self.aba1, text='Data de Emissão',bg='#dfe3ee',fg='#107db2'); self.lb_DataEmissao.place(relx=0.05,rely=0.7)
        self.DataEmissao_entry = Entry(self.aba1); self.DataEmissao_entry.place(relx=0.05,rely=0.75,relwidth=0.125)
        #  Criação da label e entrada da Data de Vencimento
        self.lb_DataVencimento = Label(self.aba1, text='Data de Vencimento',bg='#dfe3ee',fg='#107db2'); self.lb_DataVencimento.place(relx=0.2,rely=0.7)
        self.DataVencimento_entry = Entry(self.aba1); self.DataVencimento_entry.place(relx=0.2,rely=0.75,relwidth=0.15)
        #  Criação da label e entrada do Preço de Aquisição
        self.lb_PrecoAquisicao = Label(self.aba1, text='Preço de Aquisição',bg='#dfe3ee',fg='#107db2'); self.lb_PrecoAquisicao.place(relx=0.4,rely=0.7)
        self.PrecoAquisicao_entry = Entry(self.aba1); self.PrecoAquisicao_entry.place(relx=0.4,rely=0.75,relwidth=0.2)
        
    def lista_frame2(self):
        self.listaCli = ttk.Treeview(self.frame2, height=3, column=('col1','col2','col3','col4','col5',
                                                                    'col6','col7','col8','col9','col10',
                                                                    'col11','col12','col13','col14','col15',
                                                                    'col16','col17','col18','col19'))
        self.listaCli.heading('#0',text="ID")
        self.listaCli.heading('#1',text="Originador")
        self.listaCli.heading('#2',text="Doc Originador")
        self.listaCli.heading('#3',text="Cedente")
        self.listaCli.heading('#4',text="Doc Cedente")
        self.listaCli.heading('#5',text="CCB")
        self.listaCli.heading('#6',text="Cliente")
        self.listaCli.heading('#7',text="CPF/CNPJ")
        self.listaCli.heading('#8',text="Endereço")
        self.listaCli.heading('#9',text="CEP")
        self.listaCli.heading('#10',text="Cidade")
        self.listaCli.heading('#11',text="UF")
        self.listaCli.heading('#12',text="Valor do Empréstimo")
        self.listaCli.heading('#13',text="Parcela R$")
        self.listaCli.heading('#14',text="Total Parcelas")
        self.listaCli.heading('#15',text="Parcela")
        self.listaCli.heading('#16',text="Data de Emissão")
        self.listaCli.heading('#17',text="Data de Vencimento")
        self.listaCli.heading('#18',text="Preço de Aquisição")
        
        self.listaCli.column("#0",width=20)
        self.listaCli.column("#1",width=80)
        self.listaCli.column("#2",width=100)
        self.listaCli.column("#3",width=80)
        self.listaCli.column("#4",width=80)
        self.listaCli.column("#5",width=50)
        self.listaCli.column("#6",width=70)
        self.listaCli.column("#7",width=80)
        self.listaCli.column("#8",width=80)
        self.listaCli.column("#9",width=70)
        self.listaCli.column("#10",width=70)
        self.listaCli.column("#11",width=40)
        self.listaCli.column("#12",width=120)
        self.listaCli.column("#13",width=70)
        self.listaCli.column("#14",width=80)
        self.listaCli.column("#15",width=60)
        self.listaCli.column("#16",width=80)
        self.listaCli.column("#17",width=80)
        self.listaCli.column("#18",width=80)
        
        self.listaCli.place(relx=0.01,rely=0.1,relwidth=0.95,relheight=0.85)
        self.scrollLista = Scrollbar(self.frame2, orient='vertical')
        self.listaCli.configure(yscroll=self.scrollLista.set)
        self.scrollLista.place(relx=0.96, rely=0.1, relwidth=0.04, relheight=0.85)
        self.listaCli.bind("<Double-1>", self.OnDoubleClick)
        
    def Menus(self):
        menubar = Menu(self.root)
        self.root.config(menu=menubar)
        filemenu = Menu(menubar)
        filemenu2 = Menu(menubar)
        def Quit(): self.root.destroy()
        menubar.add_cascade(label='Opções',menu=filemenu)
        menubar.add_cascade(label='Sobre', menu=filemenu2)
        filemenu.add_command(label='Sair', command=Quit)
        filemenu2.add_command(label='Limpa Cliente',command=self.limpa_tela)
                            
Application()
```
